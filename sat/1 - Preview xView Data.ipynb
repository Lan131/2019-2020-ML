```python
import mxnet as mx
import gluoncv as gcv
import numpy as np
import matplotlib.pyplot as plt #this adds a lot of functionality for plotting
```

First we download the files containing the data. '.rec' and '.idx' work together so we need to download both. If you want to better understand RecordIO format please look here:
https://gluon-cv.mxnet.io/build/examples_datasets/detection_custom.html

Now we will load our datasets

```python
train = gcv.data.RecordFileDetection('../CV-Data/train_demo.rec')
val = gcv.data.RecordFileDetection('../CV-Data/val_demo.rec')
```

```python
#Classes inside of this dataset
classes = ['Passenger Vehicle',
                'Small Car',
                'Bus',
                'Pickup Truck',
                'Utility Truck',
                'Truck',
                'Cargo Truck',
                'Truck Tractor w/ Box Trailer',
                'Truck Tractor',
                'Trailer',
                'Truck Tractor w/ Flatbed Trailer',
                'Truck Tractor w/ Liquid Tank']
```

```python
i = 15 
img,lbl = train[i] 

fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(1,1,1)
gcv.utils.viz.plot_bbox(img,lbl[:,:-1],labels=lbl[:,-1],ax=ax,class_names=classes)
```

We will zoom in on the top Left

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(1,1,1)
img_small = mx.image.fixed_crop(img,0,0,328,328)
lbl_small = gcv.data.transforms.bbox.crop(lbl,(0,0,328,328))
gcv.utils.viz.plot_bbox(img_small,lbl_small[:,:-1],labels=lbl_small[:,-1],ax=ax,class_names=classes)
```

```python
### Pick your own image to view
i = 0 #set i to pick a different image


img,lbl = train[i]
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(1,1,1)
gcv.utils.viz.plot_bbox(img,lbl[:,:-1],labels=lbl[:,-1],ax=ax,class_names=classes,fontsize=8)
```

```python
x = 0 #pick your own subcrop to view x is for the x location (bottom axis)
y = 0 #pick your own subcrop to view y is for the y location (left side axis)
size = 500 # how wide/tall of a chip to view


fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(1,1,1)
img_small = mx.image.fixed_crop(img,x,y,size,size)
lbl_small = gcv.data.transforms.bbox.crop(lbl,(x,y,size,size))
gcv.utils.viz.plot_bbox(img_small,lbl_small[:,:-1],labels=lbl_small[:,-1],ax=ax,class_names=classes)
```
